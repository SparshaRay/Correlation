```python
import numpy as np
from qiskit import QuantumCircuit
from qiskit import quantum_info as qi

circ = QuantumCircuit(4)

circ.rx(np.pi/3, 0)
circ.rx(1, 2)
circ.rx(np.pi/3, 1)
circ.cx(2, 1)
circ.cx(1, 0)
circ.cx(2, 3)
circ.cx(0, 3)
circ.cx(0, 1)
circ.cx(3, 2)
circ.rx(np.pi/6, 0)
circ.ry(2*np.pi/3, 2)
circ.ry(1, 1)
circ.ry(-1, 3)
circ.h(0)
circ.h(1)
circ.h(2)
circ.h(3)

circ.draw('mpl')
```

# State before operation

```python
rho_circ = qi.DensityMatrix.from_instruction(circ)
rho_circ.draw('latex')
```
