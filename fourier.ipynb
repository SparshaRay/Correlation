```python
import numpy as np

master_list = []

def prob(qubit1, state1, qubit2 = None, state2 = None) :
    if qubit2 == None :
        return sum(state_array[state_array[:,qubit1]==state1][:,0])
    else :
        return sum(state_array[(state_array[:,qubit1]==state1) & (state_array[:,qubit2]==state2)][:,0])
    
for master in range(1000, 1005) :
    num_qubits = 20
    period = master
    state_array = []

    for j in range(2**num_qubits) :
        state_array.append([+(j%period==0), *list(map(int, [*f"{j:0{num_qubits}b}"]))])

    state_array = np.array(state_array)

    mean = prob(2, 1, 1, 1)

    master_list.append(tuple([prob(i, 1, 1, 1)-mean for i in range(3, num_qubits)]))


print(len(master_list)-len(set(master_list)))
```

```python
for elem in set(master_list) : 
    if master_list.count(elem) > 1 :
        print(master_list.index(elem))
```
